# Базовая проверка csv файлов

Простой notebook для просмотра CSV-файлов с кодировкой Windows и разделителем ';'

```python
import pandas as pd
from pathlib import Path
```

```python
data_dir = Path('../data')
csv_files = list(data_dir.rglob('*.csv'))

if csv_files:
    print(f'Найдено CSV файлов: {len(csv_files)}')
    for i, file in enumerate(csv_files):
        print(f'{i + 1}. {file}')
    
    file_path = csv_files[0]
    print(f'\nЗагружаем файл: {file_path}')
    df = pd.read_csv(file_path, encoding='cp1251', sep=';')
    print(f'Размер датафрейма: {df.shape}')
else:
    print('CSV файлы не найдены в папке data')
```

## Названия столбцов

```python
print('Столбцы в датафрейме:')
for i, col in enumerate(df.columns):
    print(f'{i+1}. {col}')
```
